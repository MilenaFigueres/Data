## COLORS

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='green')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='#FF8C00') # Usando RGB Hex Code
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='#FF8C00', linewidth=5) # que tan gruesa es la linea y se puede resumir en lw (ver mas abajo)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='#FF8C00', linewidth=5, alpha=0.5) # que tan transparente es
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='#FF8C00', lw=5, linestyle='--') # cambia el estilo de la linea
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='#FF8C00', lw=5, ls=':') # cambia el estilo de la linea y se puede poner solo ls
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='#FF8C00', lw=5, ls=':', marker='x', markersize=15)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='blue', lw=1, ls='-', marker='o', markersize=15, 
        markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='green')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=2,ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
