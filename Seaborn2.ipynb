```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.jointplot(x='fare', y='age', data=titanic, size=5)
```

```python
sns.distplot(titanic['fare'], kde=False, color='red')
```

```python
sns.boxplot(x='class', y='age', data=titanic, palette='coolwarm')
```

```python
sns.swarmplot(x='class', y='age', data=titanic, palette='coolwarm')
```

```python
sns.countplot('sex', data=titanic)
```

```python
sns.heatmap(data=titanic.corr(), cmap='coolwarm', annot=True)
```

```python
g = sns.FacetGrid(col='sex', data=titanic)
g.map(plt.hist,'age')
```
