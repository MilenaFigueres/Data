```python
sns.distplot(tips['total_bill']) #histograma con kde(linea) kernel density estimation
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips) # x e y son columnas, se comparan ambas
```

```python
tips['tip'].describe()
```

```python
 sns.jointplot(x='total_bill',y='tip',data=tips, kind='hex') #hex significa hexagon distribution  representation . Default: scatter (mirar el de arriba) 
```

```python
 sns.jointplot(x='total_bill',y='tip',data=tips, kind='reg') # otro kind
```

```python
 sns.jointplot(x='total_bill',y='tip',data=tips, kind='kde')
```

```python
sns.pairplot(tips) # pairplot va a hacer plot de parejas de relaciones a traves de toda el dataframe al menos por las columnas numericas
#seria como un jointplot pero de todas las columnas
```

```python
sns.pairplot(tips, hue='sex') #hue: the huge argument is categorical (como sex) y no numerical
# plot different color for different categorical levels using matplotlib
# hue parameter determines which column in the data frame should be used for colour encoding.
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill']) #rugplot dibuja una marca de trazos por cada punto en este uniforme o unica variante de distribucion tips['total_bill'] 
```

```python
sns.distplot(tips['total_bill']) #kde: kernel density estimation sigue una 'normal o Gaussian distribution' que es la mas comun de la distribucion de  probabilidad continua centralmente
```

```python
sns.kdeplot(tips['total_bill'])
```
